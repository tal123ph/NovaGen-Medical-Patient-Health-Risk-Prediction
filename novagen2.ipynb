# NovaGen Research Labs¶

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, recall_score,f1_score

from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression


from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import VotingClassifier
```

```python
df=pd.read_csv("novagen_dataset.csv")
df
```

```python
df.describe()
```

```python
df.info()
```

```python
x=df.drop("Target",axis=1)
y=df["Target"]


x_train,x_test,y_train,y_test=train_test_split(
    x,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y     #both sets have same proportion of target variable values
)
```

# Scale values


```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

# create your baseline model

### logistic regression with regularization

```python

lr=LogisticRegression(
    penalty="l2",
    solver="liblinear",
    max_iter=1000
)
lr.fit(x_train_scaled,y_train)
y_pred=lr.predict(x_test_scaled)

# In Model Evaluation, Recall is more important than accuracy 
# because missing a high-risk patient is dangerous

print("Logistic Regression Accuracy:", accuracy_score(y_test, y_pred))
print("Logistic Regression Recall:", recall_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

# Model 2 - KNN

```python
knn=KNeighborsClassifier(
    n_neighbors=5,
    metric="euclidean",
    n_jobs=-1
)


knn.fit(x_train_scaled,y_train)
y_pred=knn.predict(x_test_scaled)

print("KNN  Accuracy is: ", accuracy_score(y_test,y_pred)*100,"%")
print("KNN recall score is: ", recall_score(y_test,y_pred)*100,"%")
print("KNN clasification report \n: ", classification_report(y_test,y_pred))
```

## Random Forest (Ensemble Learning)

```python
rf=RandomForestClassifier(
    n_estimators=200,
    max_depth=None,
    random_state=42
)

rf.fit(x_train_scaled,y_train)
y_pred=rf.predict(x_test_scaled)

print("RandomForest Classifier accuracy is: ", accuracy_score(y_test,y_pred)*100,"%")
print("RandomForest Classifier recall is: ", recall_score(y_test,y_pred)*100,"%")

print(classification_report(y_test,y_pred))
```

## Gradient Boosting (Ensemble Learning)

```python
gb=GradientBoostingClassifier(
    n_estimators=200,
    learning_rate=0.1,
    max_depth=3,
    random_state=42
)

gb.fit(x_train_scaled,y_train)
y_pred=gb.predict(x_test_scaled)

print("Gradient Boosting Accuracy: ", accuracy_score(y_test,y_pred)*100,"%")
print("Gradient Boosting recall is: ", recall_score(y_test,y_pred)*100,"%")
print(classification_report(y_test,y_pred))

```

## Voting Classifier (Ensemble Learning)

```python
voting_cl = VotingClassifier(
    estimators=[
        ("lr", LogisticRegression(
            max_iter=1000,
            solver="liblinear"
        )),
        
        ("knn", KNeighborsClassifier(
            n_neighbors=5
        )),
        
        ("rf", RandomForestClassifier(
            n_estimators=200,
            random_state=42
        ))
    ],

    voting="soft"
)

voting_cl.fit(x_train_scaled, y_train)

y_pred = voting_cl.predict(x_test_scaled)

print("Voting Classifier Accuracy:",
      accuracy_score(y_test, y_pred) * 100, "%")

print("Voting Classifier Recall:",
      recall_score(y_test, y_pred) * 100, "%")

print(classification_report(y_test, y_pred))
```
